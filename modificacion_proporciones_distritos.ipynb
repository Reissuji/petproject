```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Dataset_Completo.csv',sep =',')
df.columns= df.columns.str.strip()
df
```

```python
df = df.drop (columns= ['Unnamed: 0'])
```

```python
df
```

```python
# cols=['Hogar','Trucos','Actividad física del dueño','Distrito']
# df[cols]=df[cols].str.strip() así no funciona

df['Hogar'] = df['Hogar'].str.strip()
df['Trucos'] = df['Trucos'].str.strip()
df['Actividad física del dueño'] = df['Actividad física del dueño'].str.strip()
df['Distrito'] = df['Distrito'].str.strip()
df['Raza'] = df['Raza'].str.strip()
df
```

```python
df['Trucos']
```

```python
# Lista con las nuevas clases y sus proporciones
nuevas_clases = [
    ("Fuencarral-El Pardo", 15),
    ("Moncloa-Aravaca", 10),
    ("Latina", 14),
    ("Carabanchel", 15),
    ("Usera", 9),
    ("Villaverde", 9),
    ("Puente de Vallecas", 18),
    ("Villa de Vallecas", 7),
    ("Moratalaz", 5),
    ("Vicálvaro", 4),
    ("Ciudad Lineal", 13),
    ("San Blas-Canillejas", 11),
    ("Hortaleza", 12),
    ("Barajas", 4),
    ("Tetuán", 9),
    ("Chamartín", 9),
    ("Chamberí", 10),
    ("Centro", 12),
    ("Retiro", 6),
    ("Arganzuela", 8),
    ("Salamanca", 10)
]

# Generar una lista con las nuevas clases replicadas según sus proporciones
nuevos_registros = []
for clase, cantidad in nuevas_clases:
    nuevos_registros.extend([clase] * cantidad)

# Mezclar aleatoriamente la lista de nuevas clases
np.random.shuffle(nuevos_registros)

# Combinar las clases existentes y nuevas
# df['barrio'] = nuevos_registros

# Reemplazar las clases existentes por las nuevas en el DataFrame original
df['Distrito'] = nuevos_registros[:210]  # Tomamos los primeros 210 registros de la lista

df

# Verificar que tenemos el número correcto de registros
# assert len(df) == 210
```

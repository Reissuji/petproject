```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv ('Top_20_perros.csv')
df
```

```python
from adjustText import adjust_text
#Importando librería adjustText para poder ajustar texto del gráfico
plt.figure(figsize=(12, 6))

# Scatter plot
plt.scatter(df['Nº Enfermedades graves'], df['Coste anual promedio, €' ], color='violet')

# Agregamos labels con el nombre de las razas para poder identificarlas en el gráfico. Con la librería adjustText evitamos que los nombres que coinciden se sobrepongan.
texts = [plt.text(x, y, raza) for x, y, raza in zip(df['Nº Enfermedades graves'],df['Coste anual promedio, €'], df['Raza'])]
adjust_text(texts, arrowprops=dict(arrowstyle='-', lw=0.7))

# Agregamos labels de los ejes y título
plt.xlabel('Nº Enfermedades graves')
plt.ylabel('Coste anual promedio, €')
plt.title('Nº Enfermedades graves vs coste anual por raza')

plt.show()
```

```python
plt.figure(figsize=(16, 10))
bar_width = 10 #defino ancho de las barras
plt.bar(df['Coste anual promedio, €' ], df['Nº Enfermedades graves'], width=bar_width, color="#00b894", edgecolor='darkgreen') #defino gráfico de barras


plt.show()
```

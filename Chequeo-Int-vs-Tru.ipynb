# Chequeo de que la relación Inteligencia vs Trucos en razas de perros está OK

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Dataset_Completo.csv')
df=df.drop('Unnamed: 0', axis=1)
df
```

## Intentando volver la inteligencia otra vez al parámetro medio de la raza, para así evitar inteligencia 0% o superiores a 100%

```python
Razas_valoresmedios = pd.read_csv('Top_20_perros.csv')
```

```python
Razas_valoresmedios = Razas_valoresmedios.drop(['Altura, cm',
                                                'Peso, kg',
                                                'Coste total, €', 
                                                'Vida promedio, años',
                                                'Nº Enfermedades graves',
                                                'Coste anual promedio, €'], axis=1)

Razas_valoresmedios
```

```python
df1 = df
df_merged = pd.merge(df, Razas_valoresmedios, on='Raza', how='left')
df1['Inteligencia (%)'] = df_merged['Inteligencia, %'].combine_first(df1['Inteligencia (%)'])
df1
```

## Viendo a ver como está la relación Inteligencia vs Trucos

```python
df1.dtypes
```

```python
df1['Trucos'].unique()
```

```python
condicion1 = df1['Inteligencia (%)'] > 80

df1[condicion1]['Trucos'].value_counts()/(len(df1[condicion1])*0.01)
```

Esta proporción parece buena, la dejo así.

En el resto del dataframe debería ser predominantemente No, así que vamos a intentar cambiarlo.

```python
condicion2 = df1['Inteligencia (%)'] < 15
condicion3 = (df1['Inteligencia (%)'] >= 15) & (df1['Inteligencia (%)'] <=40)
condicion4 = (df1['Inteligencia (%)'] > 40) & (df1['Inteligencia (%)'] <=80)

## Esto es por no cambiarlos todos y que sea como una ley física, imposible de evitar...
probabilidad_de_cambio2 = 0.9
probabilidad_de_cambio3 = 0.75
probabilidad_de_cambio4 = 0.5

n_aleatorio = np.random.rand(len(df1))

cond_cambio2 = n_aleatorio < probabilidad_de_cambio2
cond_cambio3 = n_aleatorio < probabilidad_de_cambio3
cond_cambio4 = n_aleatorio < probabilidad_de_cambio4

## Cambiar los valores de la columna destino Trucos basandonos en las condiciones y probabilidades anteriores
df1.loc[condicion2 & cond_cambio2, 'Trucos'] = 'No'
df1.loc[condicion3 & cond_cambio3, 'Trucos'] = 'No'
df1.loc[condicion4 & cond_cambio4, 'Trucos'] = 'No'
```

Chequeando que los valores están cambiados y viendo las proporciones.

Para los perritos listos:

```python
df1[condicion1]['Trucos'].value_counts()/(len(df1[condicion1])*0.01)
```

Para los perritos tontos de remate:

```python
df1[condicion2]['Trucos'].value_counts()/(len(df1[condicion2])*0.01)
```

Para los perritos de entre medias pero tontitos:

```python
df1[condicion3]['Trucos'].value_counts()/(len(df1[condicion3])*0.01)
```

```python
df1[condicion4]['Trucos'].value_counts()/(len(df1[condicion4])*0.01)
```

Se podría hacer en un momentito que hubiera más rangos, y así hacerlo más entalladito todo, pero no sé si es pasarse.

## Renombrar para evitar liarme
Antes de limpiar voy a renombrar las columnas, para así no liarme a posteriori

```python
nuevas_columnas = {
    'Dog breed':'Raza',
    'shoulder height (cm)':'Altura, cm',
    'weight (kg)':'Peso, kg',
    'INTELLIGENCE (TRAINABILITY) ranking':'Inteligencia, %',
    '4 LIFETIME COST, $':'Coste total, €',
    '2 LONGEVITY':'Vida promedio, años',
    '3 NO. OF GENETIC AILMENTS':'Nº Enfermedades graves'
}
```

NO OLVIDAR QUE TENGO QUE CAMBIAR LOS DOLARES A EUROS DESPUÉS!!!!!!

```python
df2 = df2.rename(columns=nuevas_columnas)
```

```python
df2
```

## Limpieza!

```python
df2.dtypes
```

```python
df2['Inteligencia, %']=df2['Inteligencia, %'].str.replace('%','')
df2['Coste total, €']=df2['Coste total, €'].str.replace(',','')
df2
```

```python
df2['Altura, cm'] = pd.to_numeric(df2['Altura, cm'])
df2['Peso, kg'] = pd.to_numeric(df2['Peso, kg'])
df2['Inteligencia, %'] = pd.to_numeric(df2['Inteligencia, %'])
df2['Coste total, €'] = pd.to_numeric(df2['Coste total, €'])
df2['Vida promedio, años'] = pd.to_numeric(df2['Vida promedio, años'])
df2['Nº Enfermedades graves'] = pd.to_numeric(df2['Nº Enfermedades graves'])
df2
```

```python
df2.dtypes
```

Todo correcto todo guay
### Ahora voy a cambiar de $ a € con el cambio que hay a día de hoy, Jueves 7 de Diciembre de 2023

```python
df2['Coste total, €'] = df2['Coste total, €']*0.93
df2['Coste total, €'] = df2['Coste total, €'].astype(int)
df2
```

De paso voy a crear también la columna de coste anual promedio en euros

```python
df2['Coste anual promedio, €'] = df2['Coste total, €']/df2['Vida promedio, años']
df2
```

```python
df2['Coste anual promedio, €'] = df2['Coste anual promedio, €'].astype(int)
```
